```python
df['Outlet_Size'] = df['Outlet_Size'].map({"Small":1,"Medium":2,"High":3})
#To convert string categories to integers we use the mapping.
```

```python
df.head(20)
```

```python
df['Item_Weight']=df['Item_Weight'].fillna(df['Item_Weight'].median())
#We will use the median to fill the NaN value
```

```python
df.head(20)
```

```python
df['Outlet_Size'].fillna(df['Outlet_Size'].mode()[0], inplace=True)
#We use the mode to fill the categorical values.
```
